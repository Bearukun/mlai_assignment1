```python
#Part 1.1
#Importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn 
import sklearn.linear_model
import sklearn.metrics
#Import users from json into a pandas object
data = pd.read_json("users.json")
#Let's print the data - uhmmm :) 
data
```

```python
##Plot the graph first axis is time (since registration on Hackernews) 
## and the second axis is number of points.
data.plot.scatter(x=1,y=4)
```

```python
##Part 1.2
##We chose the linear regression model because we can see a tendency that
## the number of points gets lower further along the x-axis. 
```

```python
##Part 1.3
##Now that we know what model to use, we have to train it or fit it to our data

# And we have all the x value data here: 
xs = data['created']
## Remove broken NaN-values
xs = xs[np.logical_not(np.isnan(xs))]

# And we have all the actual y value data here:
actual = data['karma']
## Remove broken NaN-values
actual = actual[np.logical_not(np.isnan(actual))]

# Reshaping the pandas dataframe to a 1D numpy array.
xs_reshape = np.array(xs).reshape(-1, 1)

# Training our model to LinearRegression
model = sklearn.linear_model.LinearRegression()

# Fit data to our model
model.fit(xs_reshape, actual)
```

```python
##Part 1.4
#1. The values are used to train our model with. These values 
# are needed to get a prediction as to how active 
# you need to be on hackernews to get high karma. 
```

```python
#2

# Make predictions using the testing set
actual_predict = model.predict(xs_reshape)

# Plot output
plt.scatter(xs, actual, color ='black')
plt.plot(xs, actual_predict, color='red', linewidth=4 )
```
